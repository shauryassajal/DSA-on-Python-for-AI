### 2. Control Structurs

#### 2.1 if-else

```python
a = -1

if a > 5:
    print("Greater than 5")
else: 
    print("Not greater than 5")
```

```python
a = 6

if a % 2 == 0: 
    print("Even")
else: 
    print("odd")
```

### 2.2 if-elif-else

```python
marks = 98

if marks > 90: 
    print('Exceptional')
elif marks > 70: 
    print('Very good')
elif marks > 50: 
    print('Good')
else: 
    print('Bad')
```

### 3. Loops

#### 3.1 For loop

```python
li = [1, 2, 3, 4, 9, 18]

print(li[0])
print(li[1])
print(li[2])
print(li[3])
print(li[4])
print(li[5])
```

```python
for element in li:
    print(element)
```

```python
for element in range(20, 31): 
    if element % 2 == 0: 
        print(element)
```

```python
for element in range(20, 31, 2):
    print(element)
```

#### 3.2 While Loop

```python
a = 1 

while a <= 5:
    print(a)
    a += 1
```
