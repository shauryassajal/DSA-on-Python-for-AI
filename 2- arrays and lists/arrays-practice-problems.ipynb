## Practice coding problems on arrays

#### 1. find the largest number in a list

```python
numbers = [12, 45, 7, 89, 34, 5, 6, 23, 2154]

# write code here
largest = numbers[0]
for element in numbers[1:]:
    if element > largest:
        largest = element
    print(f'Largest element: {largest}')
```
