## Strings Demo

### 1. strings in python

##### Ways of creating a string

```python
str1 = 'This is a string in python'
print(str1)
print(type(str1))
```

```python
str2 = "This is also a string in python"
print(str2)
print(type(str2))
```

```python
str3 = '''This is also a string with triple quotes'''
print(str3)
print(type(str3))
```

```python
str4 = """This is also a string within triple quotes"""
print(str4)
print(type(str4))
```

```python
print('Hi "John"')
```

```python
print('''this is a 
multi-line string that spans
multiple lines''')
```

```python
print('This is a \
multi-line string that spans \
multiple lines')
```
