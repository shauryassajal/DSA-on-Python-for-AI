#### 3. Common string functions

```python
str1 = 'Data SCIEnce'

print(str1)
print(id(str1))
```

***lower***

```python
lower_str1 = str1.lower()

print(lower_str1)
print(id(lower_str1))
```

```python
str1
```

***upper***

```python
str1.upper()
```

```python
str2 = 'This is 123 ABC'

print(str2.lower())
print(str2.upper())
```

***strip***

```python
str3 = '  data science  '
```

```python
str3.strip()
```

```python
str4 = '$$$###data science#####@@@'
str4.strip('$').strip('@').strip('#')
```

***replace***
