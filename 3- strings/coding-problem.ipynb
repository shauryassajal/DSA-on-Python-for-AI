##### 1. Count Vowels and Consonents
##### - assume all characters are lowercase
##### - string may include special characters

```python
s = '__hel lo 123$world '

# Write code here
vowels = 0 
consonents = 0 
for ch in s:
    if ch.isalpha():
        if ch in 'aeiou':
            vowels += 1
        else: 
            consonents += 1
print(f"There are {vowels} vowels and {consonents} consoenets")
```

#### 3. Palindrome String

```python
s = 'malayalam'

# write code here
s_rev= s[::-1]
if s == s_rev:
    print('Palindrome')
else: 
    print('Not Palindrome')
```

#### 4. Check for Anagrams
- Retun 'True' if two strings contain the same characters in any order
- Strings will be of same lenght and lowercase

```python
s1 = 'listen'
s2 = 'silent'
```

#### Time: nlog(n)

```python
s1_sorted = sorted(s1)
s2_sorted = sorted(s2)

if s1_sorted == s2_sorted :
    print('Anagram')
else: 
    print('Not Anagram')
```

```python
ord('s'), ord('s') - 97
```

```python
mapping = [0] * 26

for ch in s1: 
    index = ord(ch) - 97
    mapping[index] += 1

for ch in s2:
    index = ord(ch) - 97
    mapping[index] -= 1

if all(m == 0 for m in mapping): 
    print('Anagram')
else: 
    print('Not Anagram')
```
